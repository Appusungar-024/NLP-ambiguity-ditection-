## Notebook purpose and quick instructions

This notebook visualizes ambiguity statistics (per-genre ambiguous sentence rates) computed by
`nlp_project/compute_ambiguity_stats.py` and shows example ambiguous sentences.

If you haven't run the stats yet, run the script in a terminal first:

```bash
python3 nlp_project/compute_ambiguity_stats.py --gum-dir nlp_project/gum --pcfg nlp_project/gum.pcfg.pickle --max-per-genre 200 --topk 3
```


````python
# Imports and plotting setup
import json
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize': (10, 5)})
````

````python
# Load CSV and examples
stats_path = Path('nlp_project/ambiguity_stats.csv')
examples_path = Path('nlp_project/ambiguity_examples.json')

if not stats_path.exists():
    print('ambiguity_stats.csv not found. Run compute_ambiguity_stats.py first (see top cell).')
else:
    df = pd.read_csv(stats_path)
    display(df.sort_values('ambiguous_rate', ascending=False).head(15))

if not examples_path.exists():
    print('ambiguity_examples.json not found. Run compute_ambiguity_stats.py first (see top cell).')
else:
    with open(examples_path) as fh:
        examples = json.load(fh)
    print('Loaded examples for', len(examples), 'genres')
````

````python
# Plot ambiguous_rate by genre (if df exists)
if 'df' in globals():
    df2 = df.sort_values('ambiguous_rate', ascending=False).set_index('genre')
    ax = df2['ambiguous_rate'].plot(kind='bar')
    ax.set_ylabel('Ambiguous rate (fraction)')
    ax.set_title('Ambiguous sentences per genre (sample)')
    plt.tight_layout()
else:
    print('No data frame to plot; run compute_ambiguity_stats.py first and re-run this cell')
````

````python
# Show example ambiguous sentences for top genres
if 'examples' in globals() and examples:
    top = sorted(examples.items(), key=lambda kv: len(kv[1]), reverse=True)[:5]
    for genre, exlist in top:
        print('\n=== Genre:', genre, ' (', len(exlist), 'examples ) ===')
        for ex in exlist[:3]:
            print('- file:', ex.get('file'))
            print('  sentence:', ex.get('sentence'))
            print('  parses (sample):', ex.get('parses'))
else:
    print('No ambiguous examples available. Run compute_ambiguity_stats.py and re-run this cell')
````
